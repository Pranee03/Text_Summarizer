```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
from heapq import nlargest
###import warnings
###warnings.filterwarnings('ignore')
```

```python
text=input()
```

```python
stwds=list(STOP_WORDS)
print(stwds)
```

```python
nlp=spacy.load('en_core_web_sm')
doc=nlp(text)
tokens=[token.text for token in doc]
print(tokens)
print(punctuation)
```

```python
wd_freq={}
for word in doc:
  if word.text.lower() not in stwds:
    if word.text.lower() not in punctuation:
      if word.text not in wd_freq.keys():
        wd_freq[word.text]=1
      else:
        wd_freq[word.text]+=1
print(wd_freq)
m= max(wd_freq.values())
print(m)
```

```python
max_freq= max(wd_freq.values())
print(max_freq)
```

```python
for word in wd_freq.keys():
  wd_freq[word]=wd_freq[word]/max_freq
print(wd_freq)
```

```python
sent_tokens=[sent for sent in doc.sents]
print(sent_tokens)
```

```python
sent_src={}
for sent in sent_tokens:
  for word in sent:
    if word.text.lower() in wd_freq.keys():
      if sent not in sent_src.keys():
        sent_src[sent]=wd_freq[word.text.lower()]
      else:
        sent_src[sent]+=wd_freq[word.text.lower()]
print(sent_src)
```

```python
sel_len=int(len(sent_tokens)*0.3)
print(sel_len)
```

```python
info=nlargest(sel_len,sent_src,key=sent_src.get)
print(info)
```

```python
summary=[word.text for word in info]
info=''.join(summary)
```
